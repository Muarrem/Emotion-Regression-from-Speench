# Veri Seti
Bu kısımda veri seti okunup model eğitimi için hazırlandıktan sonra kaydetme işlemi gerçekleştirilecektir.

```python
import numpy as np
import librosa as lb
import os
```

```python
dataset_path = '/content/drive/MyDrive/Bitirme/Uygulama/Datasets/TESS/Tess'
```

## Veri setinin alınması


```python
data = []
labels = []
lenght_data = []
for dizin in os.listdir(dataset_path):
  emo_path = os.path.join(dataset_path, dizin)
  for audio_path in os.listdir(emo_path):
    data_path = os.path.join(emo_path, audio_path)
    out, sr= lb.load(data_path,sr=None)
    data.append(out)
    labels.append(str(audio_path.split("_")[2]).replace(".wav",""))
    lenght_data.append(len(out))
```

## CNN Modeli için sabit uzunluğa getirilmesi


```python
min_len = min(lenght_data)-min(lenght_data)%100
shape_data = []
for i in data:
  shape_data.append(list(i[0:min_len]))
shape_data = np.array(shape_data)
shape_data.shape
```

```python
# Veri setinin kayıt işlemi, Datasets/TESS dizininde data.npy ve labels.npy olarak kayıt edildi.
# Sonraki kodlarda bu iki dosya direk yüklenerek kullanılacaktır.
save_path = r"/content/drive/MyDrive/Bitirme/Uygulama/Datasets/TESS/"
np.save(save_path+"data.npy",shape_data)
np.save(save_path+"labels.npy",labels)
```
